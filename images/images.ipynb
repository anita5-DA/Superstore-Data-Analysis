```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 1. 'images' folder 
if not os.path.exists('images'):
    os.makedirs('images')

# 2. Data load and cleaning 
df = pd.read_csv('Sample_Cleaned.csv') 

# 3. filter of Texas's Data
texas_only = df[df['State'] == 'Texas']

# 4. Figure setup 
plt.figure(figsize=(10, 6))

# 5. Plot 
sns.barplot(data=texas_only, x='Category', y='Profit', estimator=sum, hue='Category', palette='Reds_r', legend=False)

# 6. Decoration (Title and labels)
plt.title('Profit/Loss Distribution in Texas', fontsize=15)
plt.axhline(0, color='black', linewidth=1) 

plt.savefig('images/texas_profit_loss.png', dpi=300, bbox_inches='tight')

plt.show()

print("✅ Success! The image 'texas_profit_loss.png' has been saved.")
```
